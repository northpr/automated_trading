# Momentum trading strategies with SMAs (Simple Moving Average)


```python
## Momentum trading strategies with SMAs
df.head()
```

```python
sp500 = df["Close"].to_frame()
```

```python
sp500.tail()
```

```python
sp500["sma50"] = sp500.rolling(window=50,min_periods=50).mean()
```

```python
sp500
```

```python
sp500.plot(figsize=(12,8), fontsize=15)
plt.legend(loc="upper left", fontsize=15)
```

```python
sp500["sma200"] = sp500["Close"].rolling(window=200).mean()
```

```python
sp500 = sp500.loc["2008-12-31":"2018-12-31"].copy()
```

```python
sp500.head()
```

```python
sp500.plot(figsize=(10,7))
plt.legend()
plt.grid()
```

```python
sp500.iloc[:,1:].plot(figsize=(10,7))
plt.grid()
plt.show()
```

### Conslusion - Simple Moving Average
We should invest when SMA50 is above SMA200. For example
- 2010 -> 2011, 2012 -> 2016

## EWMA

```python
sp500 = sp500["Close"].loc["2008-12-31":"2018-12-31"].copy()
```

```python
sp500 = sp500.to_frame()
```

```python
sp500.head()
```

```python
sp500["Close"].rolling(window=10).mean()
```

```python
sp500["Close"].ewm(span=10, min_periods=10).mean()
```

```python
sp500["sma"] = sp500["Close"].rolling(window=100).mean()
sp500["ema"] = sp500["Close"].ewm(span=100).mean()
```

```python
sp500.tail(100)
```

```python
sp500.iloc[500:,1:].plot(figsize=(10,7),fontsize=10)
plt.legend()
```
